## Question 1:

* Create two vectors 2dim
* Plot each vector
* Plot v1+v2, v1-v2, (v1*4+v2/2)

```python
vector1 = np.array([2, 1])
vector2 = np.array([-3, 4])

vectorSum = vector1 + vector2
vectorSubs = vector1 - vector2
vectorF = vector1 * 4 + (vector2 / 2)

##Figure Config
plt.figure(figsize=(4, 4))

plt.title("Question 1")

plt.xlim([-5, 10])
plt.ylim([-5, 10])

plt.xlabel("X")
plt.ylabel("Y")

plt.text(vector1[0], vector1[1], "v1")
plt.text(vector2[0], vector2[1], "v2")
plt.text(vectorSum[0], vectorSum[1], "vSum")
plt.text(vectorSubs[0], vectorSubs[1], "vSub")
plt.text(vectorF[0], vectorF[1], "vF")

plt.quiver(
    0,
    0,
    vector1[0],
    vector1[1],
    angles="xy",
    scale_units="xy",
    scale=1,
    color="pink",
    label="v1",
)
plt.quiver(
    0,
    0,
    vector2[0],
    vector2[1],
    angles="xy",
    scale_units="xy",
    scale=1,
    color="magenta",
    label="v2",
)
plt.quiver(
    0,
    0,
    vectorSum[0],
    vectorSum[1],
    angles="xy",
    scale_units="xy",
    scale=1,
    color="orange",
    label="vSum",
)
plt.quiver(
    0,
    0,
    vectorSubs[0],
    vectorSubs[1],
    angles="xy",
    scale_units="xy",
    scale=1,
    color="cyan",
    label="vSub",
)
plt.quiver(
    0,
    0,
    vectorF[0],
    vectorF[1],
    angles="xy",
    scale_units="xy",
    scale=1,
    color="red",
    label="vF",
)
plt.show()
```

## Question 2:

* Come up with three 2dim vectors, two of them are orthagonal, but neither is orthagonal to the third. (quiz sorusu)
* Plot all three vectors --make sure the axes are square and equal

```python
vector1 = np.array([3, -2])
vector2 = np.array([2, 3])

# To verification, dot product should be equal to zero if they're orthagonal
dot_product12 = np.dot(vector1.flatten(), vector2.flatten())
print(dot_product12)

vector3 = np.array([4, 2])
# To verification, dot product should NOT be equal to zero
dot_product31 = np.dot(vector3.flatten(), vector1.flatten())  # check if its orth. to v1
dot_product32 = np.dot(vector3.flatten(), vector2.flatten())  # check if its orth. to v2
print(dot_product31)
print(dot_product32)

##Figure Config
plt.figure(figsize=(4, 4))

plt.title("Question 2")

plt.xlim([0, 10])
plt.ylim([-5, 10])

plt.xlabel("X")
plt.ylabel("Y")

plt.text(vector1[0], vector1[1], "v1")
plt.text(vector2[0], vector2[1], "v2")
plt.text(vector3[0], vector3[1], "v3")

plt.quiver(
    0,
    0,
    vector1[0],
    vector1[1],
    angles="xy",
    scale_units="xy",
    scale=1,
    color="pink",
    label="v1",
)
plt.quiver(
    0,
    0,
    vector2[0],
    vector2[1],
    angles="xy",
    scale_units="xy",
    scale=1,
    color="magenta",
    label="v2",
)
plt.quiver(
    0,
    0,
    vector3[0],
    vector3[1],
    angles="xy",
    scale_units="xy",
    scale=1,
    color="orange",
    label="v3",
)

plt.show()
```

## Question 3:

* Create 3 matrices 2x2 2x2 3x2
* Compute scalar matrix multiplication
* Add all pairs of matrices

```python
matrix1 = np.matrix([[2, 1], [0, 1]])  # 2x2
matrix2 = np.matrix([[1, 0], [3, 4]])  # 2x2
matrix3 = np.matrix([[5, 4], [4, 2], [6, 3]])  # 3x2
s = 3  # scalar value for scalar matrix multiplication

print("matrix1*s: ", "\n", matrix1 * s, "\n")
print("matrix2*s: ", "\n", matrix2 * s, "\n")
print("matrix3*s: ", "\n", matrix3 * s, "\n")
print("matrix1+matrix2: ", "\n", np.array2string(matrix1 + matrix2), "\n")
```

## Question 4:

* What happens when transpose twice? M matrix, MT transpose, MTT
* Confirm that transpose operation works on nonsquare matrices

```python
M = np.matrix([[3, 4], [5, -6]]) #2x2 matrix
print("M:", "\n", M)
MT = M.transpose()
print("MT: ", "\n", MT)
MTT = MT.transpose()
print("MTT: ", "\n", MTT)
# as we can see, M and MTT are IDENTICAL

print("\n Confirm this on nonsquare matrices: \n")

M2=np.matrix([[1,2],[3,4],[5,6]]) #3x2 matrix
print("M2:", "\n", M2)
MT2 = M2.transpose()
print("MT2: ", "\n", MT2)
MTT2 = MT2.transpose()
print("MTT2: ", "\n", MTT2)
# M2 and MTT2 are IDENTICAL, even though M2 is a nonsquare matrix
```

## Question 5

* Create 3x3 matrix of integers
* Multiply by (a) identitiy, (b) zeros, (c) that matrix's transpose 

```python
matrixA=np.matrix([[2,1,3],[0,-3,6],[4,8,-5]])
matrixIdentity=np.eye(3)
matrixZeros=np.zeros([3,3])
matrixTranspose=matrixA.transpose()

print("matrixA: \n", matrixA)
print("\n matrixIdentity: \n", matrixIdentity)
print("\n matrixA*matrixIdentity: \n",matrixA@matrixIdentity)
print("\n matrixZeros: \n", matrixZeros)
print("\n matrixA*matrixZeros: \n",matrixA@matrixZeros)
print("\n matrixTranspose: \n", matrixTranspose)
print("\n matrixA*matrixTranspose: \n",matrixA@matrixTranspose)
```
